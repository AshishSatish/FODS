```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])


correlation_matrix = numerical_df.corr()

print(correlation_matrix)
```

## Correlation Assignment

```python
plt.figure(figsize=(10, 7))  # Changed figure size
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".1f",         # Changed format to 1 decimal point
            cmap='viridis',    # Changed color map
            square=False,      # Removed the square format
            cbar_kws={"shrink": .5, "location": "bottom"})  # Changed color bar size and location
plt.title('Correlation Matrix', fontsize=16)  # Updated title and font size
plt.xticks(rotation=45)  # Rotated x-axis labels for better readability
plt.yticks(rotation=0)   # Set y-axis labels rotation to 0
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde', corner=False)  # Plot only lower triangle, use KDE for the diagonal
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
sns.clustermap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", figsize=(10, 8))
plt.title('Clustered Correlation Heatmap')
plt.show()

```
